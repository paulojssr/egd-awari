```python
import kagglehub
import pandas as pd
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
# Fazer download do dataset Top Youtube Artists, do Kaggle.

df = pd.read_csv('./arquivos/topyoutube.csv')
```

```python
# Tratar as colunas sem informação adicionando o texto “Não informado”.

## Verificar a quantidade de valores nulos ou vazios
for coluna in df.columns:
    num_nulos = df[coluna].isna().sum()
    num_vazios = df[coluna].eq('').sum()

    print(f"Quantidade de valores nulos na coluna '{coluna}': {num_nulos}")
    print(f"Quantidade de valores vazios na coluna '{coluna}': {num_vazios}")
    print("==")
```

```python
## Substituindo valores nulos da coluna '100M'
df_format = df.fillna('Não informado')
df_format
```

```python
# Formatar coluna “AVG” para 2 casas decimais depois da vírgula.
df = df_format
df_format['Avg'] = df_format['Avg'].round(2)
df_format
```

```python
# Mostrar os top 10 usuários.
df.sort_values(by='Total Views', ascending=False).head(10)
```

```python
# Mostrar primeiros 100 usuários por nome decrescente.
df.sort_values(by='Artist', ascending=False).head(100)
```

```python
# Salvar um novo CSV com a informação tratada.
df.to_csv('./arquivos/topyoutube_format.csv', index=False)
```
